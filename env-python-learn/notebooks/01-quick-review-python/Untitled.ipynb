```python
import pandas as pd
```

```python
states = pd.read_html(
    io = "https://simple.wikipedia.org/wiki/List_of_U.S._states"    
)
```

```python
columns = [
    'state_name',
    'state_abbr',
    'capital_city',
    'largest_city',
    'established',
    'population',
    'total_area_mi',
    'total_area_km',
    'land_area_mi',
    'land_area_km',
    'water_area_mi',
    'water_area_km',
    'reps'
]

df_states.columns = columns
```
